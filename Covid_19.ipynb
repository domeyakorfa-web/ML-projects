```python
#Explore the daily vaccinations
plt.figure(figsize=(16,8))
sns.lineplot(x=df['date'],y=df['daily_vaccinations'])
plt.title('Daily vaccinations')
plt.xticks(rotation=45, ha="right") # Rotate x-axis labels
plt.show()
```

```python
countries = df.groupby('country')['total_vaccinations'].max().sort_values(ascending=False)[:5].index
print(countries)
top_countries= pd.DataFrame(columns=df.columns)
for i in countries:
  top_countries = pd.concat([top_countries, df.loc[df['country']==i]])
```

```python
plt.figure(figsize=(20,8))
sns.lineplot(data=top_countries, x='date', y='daily_vaccinations_per_million', hue='country')
plt.title('Daily vaccinations per million')
plt.xticks(rotation=45, ha="right") # Rotate x-axis labels
plt.show()
```

```python
#Question1: Which country has the most number of fully vaccinated people?
Fully_vaccinated=df.groupby('country')['people_fully_vaccinated'].max().sort_values(ascending=False).head(25)
Fully_vaccinated.reset_index()
```

```python
plt.figure(figsize=(16,10))
ax=sns.barplot(x=Fully_vaccinated.index, y=Fully_vaccinated.values)
plt.xlabel('Country')
plt.ylabel('Fully Vaccinated People')
plt.title('Top 25 Countries by Fully Vaccinated People')
plt.xticks(rotation=90) # Rotate x-axis labels for readability
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
#Question 2: Daily covid19 vaccinations per million people
Fully_vaccinated=df.groupby('country')['daily_vaccinations_per_million'].max().sort_values(ascending=False).head(25)
Fully_vaccinated.reset_index()
```

```python
plt.figure(figsize=(16,10))
daily_vax_per_million = Fully_vaccinated.reset_index()
daily_vax_per_million.columns = ['country', 'daily_vaccinations_per_million']
ax = sns.barplot(data=daily_vax_per_million,x='country',y='daily_vaccinations_per_million')
plt.xlabel('Country')
plt.ylabel('Daily Vaccinations per Million')
plt.title('Top 25 Countries by Daily Vaccinations per Million People')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
#Question 8: Find in which month there was maximum vaccinantion in any country
df['date'] = pd.to_datetime(df['date'])
df['month_year'] = df['date'].dt.to_period('M')
```

```python
monthly_vacc = (df.groupby(['country', 'month_year'])['daily_vaccinations'].sum().reset_index())
max_vacc_row = monthly_vacc.loc[monthly_vacc['daily_vaccinations'].idxmax()]
print("Country with maximum vaccinations in a single month:")
print(f"Country: {max_vacc_row['country']}")
print(f"Month: {max_vacc_row['month_year']}")
print(f"Total Vaccinations: {max_vacc_row['daily_vaccinations']:,.0f}")
```

```python
top10_months = (
    monthly_vacc.sort_values('daily_vaccinations', ascending=False)
    .head(10)
)

plt.figure(figsize=(12,6))
sns.barplot(
    data=top10_months,
    x='month_year',
    y='daily_vaccinations',
    hue='country',
    dodge=False,
    palette='coolwarm'
)
plt.title('Top 10 Highest Monthly COVID-19 Vaccination Totals by Country')
plt.xlabel('Month')
plt.ylabel('Total Vaccinations')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
